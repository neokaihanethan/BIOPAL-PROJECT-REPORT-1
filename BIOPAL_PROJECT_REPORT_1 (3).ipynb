3. Summary

This study investigates the associations between asthma, chronic bronchitis, and systemic inflammation across different ethnic groups in U.S. adults. By using six NHANES survey cycles (from 1999 to 2010), we analyse important inflammatory biomarkers, such as C Reactive Protein (CRP), urinary chloride and BMI, as well as demographic factors. Through this study, we aim to uncover disparities and patterns in correlations that can guide targeted public health strategies.

4. Background & Rationale

4.1 Health Issue or Knowledge Gap
Asthma and chronic bronchitis are prevalent respiratory conditions connected to systemic inflammation and they affect over millions of individuals in the U.S, and they are known to increase the risk of severe disease and comorbidities such as cardiovascular illnesses. While there have been previous studies that have shown that these diseases actually disproportionately affect various racial and socioeconomic groups, few studies have simultaneously analyzed their degree of association with inflammation markers across the various subgroups.
Hence, this study aims to address the knowledge gap by analysing whether inflammation related biomarkers differ between individuals with asthma and chronic bronchitis across racial and socioeconomic groups.

4.2 Public Health Importance
Chronic respiratory diseases are a major public health burden and can lead to significant morbidity, increased healthcare costs and reduced quality of life. Understanding their inflammatory mechanisms and how they differ between demographic factors is important for better disease management, and better prevention strategies as well as reduced healthcare costs. Not to mention, by identifying which subgroups cause higher inflammatory responses, we can support more precise public healthcare measures by having more tailored interventions and more efficiently allocate public health resources to cater to those which a higher risk of contracting these illnesses.

5.2 Justification of Novelty

Existing research on respiratory conditions such as asthma and chronic bronchitis often examines these diseases in isolation and analysis is typically limited to a single inflammation biomarker, such as C-reactive protein (CRP). In addition, most studies focus on a single population group, without categorising data by racial or socioeconomic subgroups, Hence due to these differences in data which are unaccounted for, it can cause critical disparities to be masked. To address this, our study adopts a multi-biomarker approach by incorporating CRP, urinary chloride, and BMI, we are able to provide a more comprehensive overview of systemic inflammation.
Additionally, our study provides a more in depth look at social inequality and how chronic diseases disproportionately affect socially disadvantaged groups such as racial/ethnic minorities and low income households, showing not only who is more likely to get sick, but also the reason behind this disproportionate effect. This also supports the increase demand in public health research to move from descriptive disparities to understanding the mechanisms involved in health inequity.

5.3 Project Gap

This study integrates three biomarkers—C-Reactive Protein, Urinary Chloride, and BMI—across six NHANES cycles to assess inflammatory in relation to respiratory diseases
While prior studies have investigated either CRP or BMI in relation to asthma or bronchitis, they often overlook the combined effects of multiple biomarkers and do not take into account and evaluate how these relationships may vary across racial or socioeconomic subgroups. In contrast we made use of the biomarker: urinary chloride — a potential indicator of acid-base balance and hydration status. However it remains an underutilized but still a physiologically relevant marker that may interact with systemic inflammation in respiratory disease. Its inclusion in our project adds a new insight to the understanding of the inflammatory profile.
This study bridges that gap by simultaneously analyzing three key biomarkers—CRP (inflammation), BMI (obesity-related inflammation), and urinary chloride (hydration/metabolic regulation)—across a wide age range of U.S. adults. It also incorporates demographic stratification by race/ethnicity and income-to-poverty ratio, allowing us to assess whether these factors mediate or modify the inflammation-respiratory disease relationship. Through this integrative approach, we aim to contribute new insights into the interdependent relationship between biological and social factors in chronic respiratory disease pathogenesis.

```python
import pandas as pd

# Mapping of variables to their source files
variable_files = {
    'RIDAGEYR': ['DEMO.xpt', 'DEMO_B.xpt', 'DEMO_C.xpt', 'DEMO_D.xpt', 'DEMO_E.xpt', 'DEMO_F.xpt'],
    'RIAGENDR': ['DEMO.xpt', 'DEMO_B.xpt', 'DEMO_C.xpt', 'DEMO_D.xpt', 'DEMO_E.xpt', 'DEMO_F.xpt'],
    'RIDRETH1': ['DEMO.xpt', 'DEMO_B.xpt', 'DEMO_C.xpt', 'DEMO_D.xpt', 'DEMO_E.xpt', 'DEMO_F.xpt'],
    'INDFMPIR': ['DEMO.xpt', 'DEMO_B.xpt', 'DEMO_C.xpt', 'DEMO_D.xpt', 'DEMO_E.xpt', 'DEMO_F.xpt'],
    'BMXBMI': ['BMX.xpt', 'BMX_B.xpt', 'BMX_C.xpt', 'BMX_D.xpt', 'BMX_E.xpt', 'BMX_F.xpt'],
    'LBXCRP': ['LAB11.xpt', 'L11_B.xpt', 'L11_C.xpt', 'CRP_D.xpt', 'CRP_E.xpt', 'CRP_F.xpt'],
    'URXUCL': ['LAB05.xpt', 'L05_B.xpt', 'L05_C.xpt', 'CHLMDA_D.xpt', 'CHLMDA_E.xpt', 'CHLMDA_F.xpt'],
    'MCQ010': ['MCQ.xpt', 'MCQ_B.xpt', 'MCQ_C.xpt', 'MCQ_D.xpt', 'MCQ_E.xpt', 'MCQ_F.xpt'],
    'MCQ160L': ['MCQ.xpt', 'MCQ_B.xpt', 'MCQ_C.xpt', 'MCQ_D.xpt', 'MCQ_E.xpt', 'MCQ_F.xpt'],
    'SMQ040': ['SMQ.xpt', 'SMQ_B.xpt', 'SMQ_C.xpt', 'SMQ_D.xpt', 'SMQ_E.xpt', 'SMQ_F.xpt'],
}

summary = []

for var, files in variable_files.items():
    combined_data = []

    for file in files:
        try:
            df = pd.read_sas(file, format='xport')
            if 'SEQN' in df.columns and var in df.columns:
                combined_data.append(df[['SEQN', var]])
        except:
            continue

    if not combined_data:
        continue

    df_all = pd.concat(combined_data, ignore_index=True)

    if var != 'RIDAGEYR':
        demo_combined = []
        for dfile in variable_files['RIDAGEYR']:
            try:
                ddf = pd.read_sas(dfile, format='xport')
                if 'SEQN' in ddf.columns and 'RIDAGEYR' in ddf.columns:
                    demo_combined.append(ddf[['SEQN', 'RIDAGEYR']])
            except:
                continue
        demo_df = pd.concat(demo_combined, ignore_index=True)
        df_all = df_all.merge(demo_df, on='SEQN', how='left')
        df_all = df_all[df_all['RIDAGEYR'].notna()]  # Ensure RIDAGEYR exists before filtering
        df_all = df_all[df_all['RIDAGEYR'] >= 20]

    total = len(df_all)
    valid = df_all[var].notna().sum()
    missing = df_all[var].isna().sum()
    missing_pct = (missing / total * 100) if total > 0 else 0

    summary.append({
        'Variable': var,
        'Total (age≥20)': total,
        'Valid': valid,
        'Missing': missing,
        'Missing %': round(missing_pct, 2)
    })

# Output final summary table
summary_df = pd.DataFrame(summary)
print(summary_df)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, shapiro, probplot

# File paths and NHANES cycles
file_info = [
    ("DEMO.xpt", "J", "1999-2000"),
    ("DEMO_B.xpt", "B", "2001–2002"),
    ("DEMO_C.xpt", "C", "2003–2004"),
    ("DEMO_D.xpt", "D", "2005–2006"),
    ("DEMO_E.xpt", "E", "2007–2008"),
    ("DEMO_F.xpt", "F", "2009–2010")
]

# Load and filter data (age ≥ 20)
frames = []
for file, cycle, year in file_info:
    df = pd.read_sas(f"{file}", format='xport', encoding='utf-8')
    if 'INDFMPIR' in df.columns and 'RIDAGEYR' in df.columns:
        df = df[['INDFMPIR', 'RIDAGEYR']].copy()
        df = df[df['RIDAGEYR'] >= 20]
        df['Cycle'] = cycle
        df['Year'] = year
        frames.append(df)

# Combine and clean
pir_df = pd.concat(frames, ignore_index=True).dropna(subset=['INDFMPIR'])

# Optional sampling
if len(pir_df) > 5000:
    pir_df = pir_df.sample(n=5000, random_state=42)

pir_data = pir_df['INDFMPIR']

# Descriptive statistics
print("Descriptive Statistics for INDFMPIR (Age ≥ 20):")
print(f"Sample Size: {len(pir_data)}")
print(f"Mean: {pir_data.mean():.2f}")
print(f"Median: {pir_data.median():.2f}")
print(f"25%: {pir_data.quantile(0.25):.2f}")
print(f"75%: {pir_data.quantile(0.75):.2f}")
print(f"Min: {pir_data.min():.2f}")
print(f"Max: {pir_data.max():.2f}")
print(f"Skewness: {skew(pir_data):.3f}")

# Normality test
shapiro_stat, shapiro_p = shapiro(pir_data)
print(f"Shapiro-Wilk p-value: {shapiro_p:.4f}")
if shapiro_p < 0.05:
    print("Interpretation: Not normally distributed (p < 0.05).")
else:
    print("Interpretation: Normally distributed.")

# Plot
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(pir_data, bins=40, kde=True, color='purple')
plt.title('Histogram of INDFMPIR (Age ≥ 20)')
plt.xlabel('Income to Poverty Ratio')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
probplot(pir_data, dist="norm", plot=plt)
plt.title('Q-Q Plot of INDFMPIR')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, shapiro, probplot

# Step 1: File paths and cycle metadata for RIDAGEYR
file_info = [
    ("DEMO.xpt", "J", "1999–2000"),
    ("DEMO_B.xpt", "B", "2001–2002"),
    ("DEMO_C.xpt", "C", "2003–2004"),
    ("DEMO_D.xpt", "D", "2005–2006"),
    ("DEMO_E.xpt", "E", "2007–2008"),
    ("DEMO_F.xpt", "F", "2009–2010")
]

# Step 2: Load and filter age ≥ 20
frames = []
for file, cycle, year in file_info:
    df = pd.read_sas(f"{file}", format='xport', encoding='utf-8')
    if 'RIDAGEYR' in df.columns:
        df = df[['RIDAGEYR']].copy()
        df = df[df['RIDAGEYR'] >= 20]
        df['Cycle'] = cycle
        df['Year'] = year
        frames.append(df)

# Step 3: Combine and clean
age_df = pd.concat(frames, ignore_index=True).dropna(subset=['RIDAGEYR'])

# Step 4: Sample if necessary
if len(age_df) > 5000:
    age_df = age_df.sample(n=5000, random_state=42)

age_data = age_df['RIDAGEYR']

# Step 5: Descriptive statistics
print("Descriptive Statistics for RIDAGEYR (Age ≥ 20):")
print(f"Sample Size: {len(age_data)}")
print(f"Mean: {age_data.mean():.2f}")
print(f"Median: {age_data.median():.2f}")
print(f"25%: {age_data.quantile(0.25):.2f}")
print(f"75%: {age_data.quantile(0.75):.2f}")
print(f"Min: {age_data.min():.2f}")
print(f"Max: {age_data.max():.2f}")

# Step 5.1: Skewness interpretation
skew_val = skew(age_data)
print(f"Skewness: {skew_val:.3f}")
if skew_val <= -1:
    print("Interpretation: Strong left-skew (tail toward younger adults).")
elif -1 < skew_val <= -0.5:
    print("Interpretation: Moderate left-skew.")
elif -0.5 < skew_val < 0.5:
    print("Interpretation: Approximately symmetric distribution.")
elif 0.5 <= skew_val < 1:
    print("Interpretation: Moderate right-skew (tail toward older adults).")
else:
    print("Interpretation: Strong right-skew (fewer very old participants).")

# Step 6: Shapiro-Wilk test
shapiro_stat, shapiro_p = shapiro(age_data)
print(f"Shapiro-Wilk p-value (n={len(age_data)}): {shapiro_p:.4f}")
if shapiro_p < 0.05:
    print("Interpretation: Data is likely not normally distributed (p < 0.05).")
else:
    print("Interpretation: Data does not significantly differ from a normal distribution.")

# Step 7: Visualizations
plt.figure(figsize=(12, 5))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(age_data, bins=40, kde=True, color='lightgreen')
plt.title('Histogram of RIDAGEYR (Age ≥ 20)')
plt.xlabel('Age (Years)')
plt.ylabel('Frequency')

# Q-Q Plot
plt.subplot(1, 2, 2)
probplot(age_data, dist="norm", plot=plt)
plt.title('Q-Q Plot of RIDAGEYR')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, shapiro, probplot

# Step 1: File paths and metadata
bmx_files = [
    ("BMX.xpt", "1999–2000"),
    ("BMX_B.xpt", "2001–2002"),
    ("BMX_C.xpt", "2003–2004"),
    ("BMX_D.xpt", "2005–2006"),
    ("BMX_E.xpt", "2007–2008"),
    ("BMX_F.xpt", "2009–2010")
]

demo_files = [
    ("DEMO.xpt", "1999–2000"),
    ("DEMO_B.xpt", "2001–2002"),
    ("DEMO_C.xpt", "2003–2004"),
    ("DEMO_D.xpt", "2005–2006"),
    ("DEMO_E.xpt", "2007–2008"),
    ("DEMO_F.xpt", "2009–2010")
]

# Step 2: Load, merge BMXBMI with age, and filter age ≥ 20
frames = []
for (bmx_file, year), (demo_file, _) in zip(bmx_files, demo_files):
    bmx = pd.read_sas(bmx_file, format='xport')
    demo = pd.read_sas(demo_file, format='xport')

    if 'SEQN' in bmx.columns and 'SEQN' in demo.columns:
        merged = pd.merge(bmx, demo[['SEQN', 'RIDAGEYR']], on='SEQN', how='inner')
        merged = merged[['BMXBMI', 'RIDAGEYR']].dropna()
        merged = merged[merged['RIDAGEYR'] >= 20]
        merged['Year'] = year
        frames.append(merged)

# Step 3: Combine all
bmi_df = pd.concat(frames, ignore_index=True)

# Step 4: Sample if needed
if len(bmi_df) > 5000:
    bmi_df = bmi_df.sample(n=5000, random_state=42)

bmi_data = bmi_df['BMXBMI']

# Step 5: Descriptive stats
print("Descriptive Statistics for BMXBMI (Age ≥ 20):")
print(f"Sample Size: {len(bmi_data)}")
print(f"Mean: {bmi_data.mean():.2f}")
print(f"Median: {bmi_data.median():.2f}")
print(f"25%: {bmi_data.quantile(0.25):.2f}")
print(f"75%: {bmi_data.quantile(0.75):.2f}")
print(f"Min: {bmi_data.min():.2f}")
print(f"Max: {bmi_data.max():.2f}")

# Step 5.1: Skewness
skew_val = skew(bmi_data)
print(f"Skewness: {skew_val:.3f}")
if skew_val <= -1:
    print("Interpretation: Strong left-skew")
elif -1 < skew_val <= -0.5:
    print("Interpretation: Moderate left-skew")
elif -0.5 < skew_val < 0.5:
    print("Interpretation: Approximately symmetric")
elif 0.5 <= skew_val < 1:
    print("Interpretation: Moderate right-skew")
else:
    print("Interpretation: Strong right-skew")

# Step 6: Shapiro-Wilk test
shapiro_stat, shapiro_p = shapiro(bmi_data)
print(f"Shapiro-Wilk p-value: {shapiro_p:.4f}")
if shapiro_p < 0.05:
    print("Interpretation: Not normally distributed")
else:
    print("Interpretation: Data is approximately normal")

# Step 7: Visualization
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(bmi_data, bins=40, kde=True, color='orange')
plt.title('Histogram of BMXBMI (Age ≥ 20)')
plt.xlabel('BMI (kg/m²)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
probplot(bmi_data, dist="norm", plot=plt)
plt.title('Q-Q Plot of BMXBMI')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, shapiro, probplot

# File paths and metadata
ucl_file_info = [
    ("LAB05.xpt", "J", "1999–2000"), # Corrected filename
    ("L05_B.xpt", "B", "2001–2002"),
    ("L05_C.xpt", "C", "2003–2004"),
    ("CHLMDA_D.xpt", "D", "2005–2006"),
    ("CHLMDA_E.xpt", "E", "2007–2008"),
    ("CHLMDA_F.xpt", "F", "2009–2010")
]

demo_file_info = [
    ("DEMO.xpt", "J", "1999–2000"), # Added cycle and year for consistency, corrected path
    ("DEMO_B.xpt", "B", "2001–2002"), # Corrected path
    ("DEMO_C.xpt", "C", "2003–2004"), # Corrected path
    ("DEMO_D.xpt", "D", "2005–2006"), # Corrected path
    ("DEMO_E.xpt", "E", "2007–2008"), # Corrected path
    ("DEMO_F.xpt", "F", "2009–2010")  # Corrected path
]

# Load and combine URXUCL and RIDAGEYR for age filtering
frames = []
for i, (ucl_file, cycle, year) in enumerate(ucl_file_info):
    # Use the correct demo file based on index
    demo_file, _, _ = demo_file_info[i]
    ucl_df = pd.read_sas(ucl_file, format='xport', encoding='utf-8')
    demo_df = pd.read_sas(demo_file, format='xport', encoding='utf-8')


    if 'SEQN' in ucl_df.columns and 'SEQN' in demo_df.columns and 'RIDAGEYR' in demo_df.columns and 'URXUCL' in ucl_df.columns:
        merged = pd.merge(ucl_df[['SEQN', 'URXUCL']], demo_df[['SEQN', 'RIDAGEYR']], on='SEQN')
        merged['Cycle'] = cycle
        merged['Year'] = year
        frames.append(merged)
    else:
      print(f"Missing required columns in {ucl_file} or {demo_file}")


# Combine all
ucl_all = pd.concat(frames, ignore_index=True)

# Filter for participants aged 20+
ucl_filtered = ucl_all[(ucl_all['RIDAGEYR'] >= 20) & (~ucl_all['URXUCL'].isna())].dropna(subset=['URXUCL']) # Ensure URXUCL is not NaN after filtering

# Sample if needed
if len(ucl_filtered) > 5000:
    ucl_filtered = ucl_filtered.sample(n=5000, random_state=42)

ucl_data = ucl_filtered['URXUCL']

# Descriptive statistics
print("Descriptive Statistics for URXUCL (Urinary Chloride) [Age ≥ 20]:")
print(f"Sample Size: {len(ucl_data)}")
print(f"Mean: {ucl_data.mean():.2f}")
print(f"Median: {ucl_data.median():.2f}")
print(f"25%: {ucl_data.quantile(0.25):.2f}")
print(f"75%: {ucl_data.quantile(0.75):.2f}")
print(f"Min: {ucl_data.min():.2f}")
print(f"Max: {ucl_data.max():.2f}")
skew_val = skew(ucl_data)
print(f"Skewness: {skew_val:.3f}")

if skew_val < -1:
    print("Interpretation: Strong left-skew.")
elif skew_val > 1:
    print("Interpretation: Strong right-skew.")
else:
    print("Interpretation: Distribution is approximately symmetric.")

# Shapiro-Wilk Test
shapiro_stat, shapiro_p = shapiro(ucl_data)
print(f"Shapiro-Wilk p-value: {shapiro_p:.4f}")
if shapiro_p < 0.05:
    print("Interpretation: Data likely not normally distributed.")
else:
    print("Interpretation: Data does not significantly differ from normal.")

# Visuals
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(ucl_data, bins=10, kde=True, color='green')
plt.title('Histogram of URXUCL (Urinary Chloride)')
plt.xlabel('Urinary Chloride (mg/L)')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
probplot(ucl_data, dist="norm", plot=plt)
plt.title('Q-Q Plot of URXUCL')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Define file paths and corresponding variable origins
file_map = {
    "DEMO": {
        "files": [
            "DEMO.xpt", "DEMO_B.xpt", "DEMO_C.xpt",
            "DEMO_D.xpt", "DEMO_E.xpt", "DEMO_F.xpt"
        ],
        "variables": ["RIDAGEYR", "RIAGENDR", "RIDRETH1", "INDFMPIR"]
    },
    "MCQ": {
        "files": [
            "MCQ.xpt", "MCQ_B.xpt", "MCQ_C.xpt",
            "MCQ_D.xpt", "MCQ_E.xpt", "MCQ_F.xpt"
        ],
        "variables": ["MCQ010", "MCQ160L"]
    },
    "SMQ": {
        "files": [
            "SMQ.xpt", "SMQ_B.xpt", "SMQ_C.xpt",
            "SMQ_D.xpt", "SMQ_E.xpt", "SMQ_F.xpt"
        ],
        "variables": ["SMQ040"]
    },
    "BMX": {
        "files": [
            "BMX.xpt", "BMX_B.xpt", "BMX_C.xpt",
            "BMX_D.xpt", "BMX_E.xpt", "BMX_F.xpt"
        ],
        "variables": ["BMXBMI"]
    }
}

# Load and merge data
merged_data = []

for i in range(6):  # 6 NHANES cycles
    dfs = []
    for domain, details in file_map.items():
        file_path = details["files"][i]
        try:
            df = pd.read_sas(file_path, format='xport', encoding='utf-8')
            df = df[['SEQN'] + details["variables"]]
            dfs.append(df)
        except FileNotFoundError as e:
            print(f"File not found: {file_path}. Skipping.")
            continue
        except Exception as e:
            print(f"Error processing file {file_path}: {e}. Skipping.")
            continue

    if dfs:
        df_merged = dfs[0]
        for df in dfs[1:]:
            df_merged = df_merged.merge(df, on='SEQN', how='outer')
        merged_data.append(df_merged)

# Combine all cycles
if merged_data:
    final_df = pd.concat(merged_data, ignore_index=True)

    # Filter to adults age 20+
    final_df = final_df[final_df['RIDAGEYR'] >= 20]

    # Drop rows with missing BMI
    final_df = final_df.dropna(subset=['BMXBMI'])

    # Categorize variables
    final_df['Asthma'] = final_df['MCQ010'].map({1: 'Yes', 2: 'No'})
    final_df['Chronic Bronchitis'] = final_df['MCQ160L'].map({1: 'Yes', 2: 'No'})
    final_df['Gender'] = final_df['RIAGENDR'].map({1: 'Male', 2: 'Female'})
    final_df['Race/Ethnicity'] = final_df['RIDRETH1'].map({
        1: 'Mexican American',
        2: 'Other Hispanic',
        3: 'Non-Hispanic White',
        4: 'Non-Hispanic Black',
        5: 'Other/Multi-Racial'
    })
    final_df['Smoking'] = final_df['SMQ040'].map({1: 'Everyday', 2: 'Some days', 3: 'Not at all'})

    # Age groups
    final_df['Age Group'] = pd.cut(final_df['RIDAGEYR'], bins=[19, 40, 60, 120], labels=['20–40', '40–60', '60+'])

    # Income groups
    final_df['Income Group'] = pd.cut(final_df['INDFMPIR'], bins=[-1, 1.3, 3.5, final_df['INDFMPIR'].max()],
                                      labels=['Low', 'Medium', 'High'])

    # Melt for plotting
    melted_df = pd.melt(
        final_df,
        id_vars=['BMXBMI'],
        value_vars=['Asthma', 'Chronic Bronchitis', 'Race/Ethnicity', 'Income Group', 'Gender', 'Age Group', 'Smoking'],
        var_name='Independent Variable',
        value_name='Category'
    )

    # Compute means
    plot_data = melted_df.groupby(['Independent Variable', 'Category'], observed=True)['BMXBMI'].mean().reset_index()

    # Plot
    plt.figure(figsize=(16, 10))
    sns.barplot(data=plot_data, x='Category', y='BMXBMI', hue='Independent Variable', width=1.0)
    plt.xticks(rotation=45)
    plt.title('Effect of Independent Variables on BMI (Adults 20+)')
    plt.ylabel('Mean BMI')
    plt.tight_layout()
    plt.show()
else:
    print("No data was loaded. Please check the file paths and ensure the necessary files are available.")
```

The graph shows that BMI varies meaningfully across demographic, clinical, and behavioral subgroups. The most notable disparities are linked to:

Income (higher BMI in lower-income adults),

Asthma presence,

and Race/Ethnicity.

These patterns support the research question by showing how BMI (a marker of systemic inflammation and chronic disease risk) is stratified across racial and socioeconomic groups, and potentially linked to asthma and bronchitis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Define actual uploaded file paths - removed /mnt/data/ prefix
file_cycles_actual = {
    '1999–2000': {
        'DEMO': 'DEMO.xpt',
        'MCQ': 'MCQ.xpt'
    },
    '2001–2002': {
        'DEMO': 'DEMO_B.xpt',
        'MCQ': 'MCQ_B.xpt'
    },
    '2003–2004': {
        'DEMO': 'DEMO_C.xpt',
        'MCQ': 'MCQ_C.xpt'
    },
    '2005–2006': {
        'DEMO': 'DEMO_D.xpt',
        'MCQ': 'MCQ_D.xpt'
    },
    '2007–2008': {
        'DEMO': 'DEMO_E.xpt',
        'MCQ': 'MCQ_E.xpt'
    },
    '2009–2010': {
        'DEMO': 'DEMO_F.xpt',
        'MCQ': 'MCQ_F.xpt'
    }
}

asthma_bronchitis_stats = []

for cycle, files in file_cycles_actual.items():
    try:
        demo = pd.read_sas(files['DEMO'], format='xport')[['SEQN', 'RIDAGEYR']]
        mcq = pd.read_sas(files['MCQ'], format='xport')[['SEQN', 'MCQ010', 'MCQ160L']]

        merged = demo.merge(mcq, on='SEQN', how='inner')
        merged = merged[merged["RIDAGEYR"] >= 20]

        asthma_pct = (merged['MCQ010'] == 1).mean() * 100
        bronchitis_pct = (merged['MCQ160L'] == 1).mean() * 100

        asthma_bronchitis_stats.append({
            'Cycle': cycle,
            'Asthma_%': asthma_pct,
            'Bronchitis_%': bronchitis_pct
        })

    except FileNotFoundError as e:
        print(f"Error processing {cycle}: {e}. Skipping cycle.")
    except Exception as e:
        print(f"An unexpected error occurred while processing {cycle}: {e}. Skipping cycle.")


diagnosis_df = pd.DataFrame(asthma_bronchitis_stats)

# Only attempt to plot if there is data in the DataFrame
if not diagnosis_df.empty:
    plt.figure(figsize=(12, 6))
    plt.plot(diagnosis_df['Cycle'], diagnosis_df['Asthma_%'], marker='o', label='Asthma')
    plt.plot(diagnosis_df['Cycle'], diagnosis_df['Bronchitis_%'], marker='o', label='Chronic Bronchitis')
    plt.title('Percentage of Adults (20+) Diagnosed with Asthma or Chronic Bronchitis')
    plt.xlabel('NHANES Cycle')
    plt.ylabel('Percentage Diagnosed')
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No data was loaded to plot. Please check file names and availability.")
```

Asthma Diagnosis:
The percentage of U.S. adults reporting a physician-diagnosed asthma condition has increased steadily from 1999–2000 to 2009–2010, rising from around 10.5% to 13.6%. This suggests that asthma is becoming a more common chronic condition among adults. This trend may be associated with increasing rates of obesity and systemic inflammation (as seen in rising BMI and CRP levels), both of which are known to exacerbate or contribute to respiratory conditions.

Chronic Bronchitis Diagnosis:
The percentage of adults diagnosed with chronic bronchitis remained relatively stable over the same period, fluctuating slightly between 2.9% and 3.8%. Unlike asthma, chronic bronchitis did not show a clear upward trend. This stability might be influenced by improvements in public health policies (e.g., smoking cessation campaigns), diagnostic practices, or potential underreporting.

The rising asthma prevalence alongside increasing CRP and BMI levels supports a potential association between metabolic/inflammatory factors and asthma development or worsening.
